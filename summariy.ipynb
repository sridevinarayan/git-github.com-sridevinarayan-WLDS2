```python
#read the results
import numpy as np 
import pandas as pd

# normalized_out = '/Users/ns/development/iisc/WLDS2/normal_out/plots_1.csv'
# un_normalized_out = '/Users/ns/development/iisc/WLDS2/un_normal_out/plots_1.csv'
# normal_df = pd.read_csv(normalized_out)
# un_normal_df = pd.read_csv(normalized_out)
# normal_df['data_normalized'] = 'normalized'
# un_normal_df['data_normalized'] = 'not normalized'
# # normal_df.head()
# un_normal_df.head()

#*** 
#Metrics
norm_metrics = '/Users/ns/development/iisc/WLDS2/normal_out/metrics_1.csv'
un_norm_metrics = '/Users/ns/development/iisc/WLDS2/un_normal_out/metrics_1.csv'
normal_m_df = pd.read_csv(norm_metrics)
un_normal_m_df = pd.read_csv(un_norm_metrics)
normal_m_df['data_normalized'] = 'normalized'
un_normal_m_df['data_normalized'] = 'not normalized'
# un_normal_m_df.head()
# normal_m_df.head()

concat_data_df = pd.concat([normal_m_df, un_normal_m_df], ignore_index=True)
concat_data_df[['test_set', 'tr_epoch', 'data_normalized', 'label', 'pred']]
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from ast import literal_eval

test_score = concat_data_df['test_score']
#print(test_score[0])
list_acc=[]
list_loss=[]
for index, row in concat_data_df.iterrows():
        li = literal_eval(row['test_score'])
        list_acc.append(li[1])
        list_loss.append(li[0])
        # print(li[0])
        # print(li[1])
        # print(li)
concat_data_df['val_acc'] = list_acc
concat_data_df['val_loss'] = list_loss
#concat_data_df.head()
g = sns.relplot(x="tr_epoch", y="val_acc", hue="data_normalized",
                col='test_set', col_wrap=3, data=concat_data_df, kind='line')
g.figure.suptitle("Validation Accuracy for test splits",
                          fontsize=20, fontweight='bold')
g.set_axis_labels("#Epoch", "Validation Accuracy", size=12, weight='bold')
g.figure.subplots_adjust(hspace=.7, top=.80)


g1 = sns.relplot(x="tr_epoch", y="val_loss", hue="data_normalized",
                col='test_set', col_wrap=3, data=concat_data_df, kind='line')
g1.figure.suptitle("Validation loss for test splits",
                  fontsize=20, fontweight='bold')
g1.set_axis_labels("#Epoch", "Validation Loss", size=12, weight='bold')
g1.figure.subplots_adjust(hspace=.7, top=.80)
```
